## Hierarchical Clustering

### Introdução

O Hierarchical Clustering, ou Agrupamento Hierárquico, é um método de agrupamento que constrói uma hierarquia de clusters. Ao contrário de abordagens que definem um número fixo de clusters, o Hierarchical Clustering organiza os dados em uma estrutura de árvore, permitindo a análise de diferentes níveis de granularidade. Existem duas abordagens principais: o agrupamento aglomerativo, onde os dados são inicialmente tratados como pontos individuais e, em seguida, agrupados iterativamente, e o agrupamento divisivo, que começa com um cluster único e divide-o progressivamente.



### Aplicações

- Biologia Molecular: No campo da biologia molecular, o Hierarchical Clustering é frequentemente utilizado para analisar perfis de expressão genética, ajudando a identificar padrões de regulação gênica e relações entre diferentes amostras biológicas.

- Análise de Mercado: Em análise de mercado, o Hierarchical Clustering pode ser aplicado para segmentar clientes com base em comportamentos de compra, permitindo estratégias de marketing mais direcionadas.

- Sensoriamento Remoto: Na interpretação de dados de sensoriamento remoto, o Hierarchical Clustering pode agrupar pixels semelhantes em imagens, facilitando a identificação de padrões geoespaciais.

### Vantagens

1. Interpretabilidade: A estrutura hierárquica proporciona uma visão intuitiva da organização dos dados, permitindo a interpretação em diferentes níveis de detalhe.

2. Robustez a Outliers: O Hierarchical Clustering tende a ser robusto em relação a outliers, já que a estrutura de árvore suaviza a influência de pontos atípicos.

3. Não Requer Especificação Prévia de Número de Clusters: Ao contrário de métodos que exigem a definição prévia do número de clusters, o Hierarchical Clustering não requer essa informação, tornando-o mais flexível.

### Desvantagens

1. Custo Computacional Elevado: O processo de construção da árvore hierárquica pode ser computacionalmente intensivo, especialmente em conjuntos de dados grandes.

2. Sensibilidade à Métrica de Similaridade: A escolha da métrica de similaridade pode influenciar significativamente os resultados, exigindo uma seleção cuidadosa para garantir resultados robustos.

3. Dificuldade com Conjuntos de Dados Não-Esfericamente Distribuídos: O Hierarchical Clustering pode enfrentar desafios ao lidar com conjuntos de dados não-esfericamente distribuídos, onde clusters têm formas complexas.

### Exemplo de código

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Gerando dados de exemplo
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.40)
```

```python
# Aplicando o Hierarchical Clustering Aglomerativo
agglomerative_cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
labels = agglomerative_cluster.fit_predict(X)
```

```python
# Visualizando os resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('Hierarchical Clustering')
plt.show()
```

```python
# Criando um dendrograma para visualizar a hierarquia
linked = linkage(X, 'ward')
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrograma Hierárquico')
plt.show()
```
